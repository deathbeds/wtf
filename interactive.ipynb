# Prototypical Essay: Hypothesis Testing

- ### Define a question
- ### Gather information and resources
- ### Form an explanatory hypothesis
- ### Test the hypothesis by performing an experiment \nand collecting data in a reproducible manner
- ### Analyze the data
- ### Interpret the data and draw conclusions that serve \nas a starting point for new hypothesis
- ### Publish results
- ### Retest
